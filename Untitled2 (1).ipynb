Q1. What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Ans.
 Anomaly Detection refers to the process of identifying unusual patterns, data points, or events that deviate significantly from the majority of observations in a dataset. These anomalies often indicate critical information such as fraud, system failures, or unexpected events. For example, in banking transactions, a sudden high-value withdrawal may be an anomaly indicating fraud.
There are three main types of anomalies:
1. Point Anomalies: These are individual data points that stand out from the rest. Example: A credit card transaction of ₹1,00,000 when the usual spending is around ₹2,000.


2. Contextual Anomalies: These depend on context, such as time or location. Example: A temperature of 30°C may be normal in Delhi summer but abnormal in Shimla winter.


3. Collective Anomalies: A group of related data points that deviate when observed together. Example: In network security, a sudden burst of traffic from multiple IPs may indicate a Distributed Denial of Service (DDoS) attack.


In real-world business, anomaly detection helps in fraud detection, healthcare monitoring, manufacturing quality control, and cybersecurity.

Q2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Ans.
1. Isolation Forest (iForest):


* Approach: Randomly partitions the data using decision trees. Anomalies are isolated quickly since they require fewer splits.


* Strengths: Works well with high-dimensional data, scalable for large datasets.


* Use Case: Fraud detection in financial transactions, server log anomaly detection.


2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise):


* Approach: Groups data points into clusters based on density. Points that lie in low-density regions are labeled anomalies.


* Strengths: Can detect arbitrarily shaped clusters and noise.


* Use Case: Spatial/geographical data analysis, customer segmentation, detecting noise in GPS data.


3. Local Outlier Factor (LOF):


* Approach: Compares the local density of a point with its neighbors. Points with significantly lower density are anomalies.


* Strengths: Captures local patterns better than global models.

* Use Case: Detecting fraudulent user behavior in e-commerce, medical outlier detection in patient vitals.

Q3. What are the key components of a Time Series? Explain each with one example.

Ans.
 A time series is a sequence of data points collected over time. The key components are:
1. Trend: The long-term movement in data. Example: The steady increase in India’s smartphone usage over the last decade.


2. Seasonality: Regular patterns that repeat at fixed intervals. Example: Ice cream sales peak during summer months.


3. Cyclic Patterns: Long-term fluctuations not tied to fixed seasons. Example: Business cycles like boom and recession phases in the economy.


4. Irregular/Random Component: Unpredictable fluctuations caused by random events. Example: A sudden drop in airline passengers during COVID-19 lockdowns.


Q4. Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Ans.
 A stationary time series is one where the statistical properties such as mean, variance, and autocorrelation remain constant over time. Stationarity is crucial for time series modeling because most models like ARIMA assume that the data is stationary.

Testing Stationarity:
1. Visual Inspection: Plotting the series to check for constant mean and variance.


2. Summary Statistics: Dividing data into segments and comparing means/variances.


3. Statistical Tests: Augmented Dickey-Fuller (ADF) and KPSS tests are commonly used.


Transforming Non-Stationary Series:
1. Differencing: Subtracting current values from previous ones to remove trend.


2. Log Transformation: Stabilizes variance in datasets like financial time series.


3. Detrending/Decomposition: Separating trend and seasonality components.


4. Box-Cox Transformation: Used for stabilizing variance in skewed datasets.


Example: The AirPassengers dataset (monthly airline passengers 1949–1960) is non-stationary due to increasing trend and variance. Applying log transformation and differencing makes it stationary, enabling ARIMA/SARIMA modeling.


Q5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

Ans.
1. AR (AutoRegressive) Model:


* Structure: Predicts future values using linear combinations of past values.


* Use Case: Stock prices where past values influence current prices.


2. MA (Moving Average) Model:


* Structure: Uses past forecast errors to predict future values.


* Use Case: Weather forecast adjustments using residuals.


3. ARIMA (AutoRegressive Integrated Moving Average):


* Structure: Combines AR and MA with differencing (I) to handle non-stationary series.


* Use Case: Forecasting non-seasonal data such as GDP growth.


4. SARIMA (Seasonal ARIMA):


* Structure: Extends ARIMA by adding seasonal components.


* Use Case: Airline passenger demand where seasonality plays a role.


5. SARIMAX (Seasonal ARIMA with Exogenous Variables):


* Structure: SARIMA + external regressors (e.g., weather, holidays).


* Use Case: Energy demand forecasting influenced by temperature.

```python
#Q7. Apply Isolation Forest on NYC Taxi Fare dataset to detect anomalies and visualize them

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from google.colab import drive

# Mount Drive
drive.mount('/content/drive')

# Load NYC Taxi Fare dataset
nyc_taxi_path = "/content/drive/MyDrive/Copy of NYC_taxi_fare_data.csv"
taxi_df = pd.read_csv(nyc_taxi_path)

# Select relevant columns (adjust if your file has different names)
df = taxi_df[['fare_amount','tip_amount']].dropna()

# Apply Isolation Forest
clf = IsolationForest(contamination=0.02, random_state=42)
df['anomaly'] = clf.fit_predict(df[['fare_amount','tip_amount']])

# Visualize anomalies
plt.figure(figsize=(8,6))
plt.scatter(df['fare_amount'], df['tip_amount'], c=df['anomaly'], cmap="coolwarm", s=10)
plt.xlabel("Fare Amount")
plt.ylabel("Tip Amount")
plt.title("Isolation Forest - NYC Taxi Fare Anomalies")
plt.show()
```

```python
#Q8. Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from google.colab import drive

# Mount Drive
drive.mount('/content/drive')

# Load AirPassengers dataset
air_passengers_path = "/content/drive/MyDrive/Copy of AirPassengers.csv"
air_df = pd.read_csv(air_passengers_path)

# Convert Month column to datetime and set as index
air_df['Month'] = pd.to_datetime(air_df['Month'])
air_df.set_index('Month', inplace=True)

# Train SARIMA model
model = SARIMAX(air_df['#Passengers'], order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit(disp=False)

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_index = pd.date_range(start=air_df.index[-1] + pd.offsets.MonthBegin(1), periods=12, freq="M")

# Plot results
plt.figure(figsize=(10,5))
plt.plot(air_df['#Passengers'], label="Observed")
plt.plot(forecast_index, forecast.predicted_mean, label="Forecast", color="red")
plt.fill_between(forecast_index,
                 forecast.conf_int()["lower #Passengers"],
                 forecast.conf_int()["upper #Passengers"],
                 color="pink", alpha=0.3)
plt.title("SARIMA Forecast of AirPassengers")
plt.legend()
plt.show()
```

Q10. You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.

Ans.

🔹 1. Anomaly Detection in Streaming Data

For real-time anomaly detection, we can use algorithms that handle incremental data:

* Isolation Forest: Efficient for high-dimensional numerical data, can quickly detect spikes/drops in usage.

* Local Outlier Factor (LOF): Good for detecting local density anomalies, such as sudden drops in one region compared to its neighbors.

* DBSCAN: Useful if the anomalies form clusters (e.g., many households in one region suddenly drop usage).

For streaming, these methods can be applied in a sliding window fashion, where the last few hours/days of data are continuously monitored.

🔹 2. Short-Term Forecasting Model

For forecasting energy demand:

* ARIMA works well for stationary series but ignores seasonality.

* SARIMA handles daily/weekly/seasonal patterns in electricity demand.

* SARIMAX is best here because external factors (temperature, humidity, holidays) strongly affect energy usage.

Example: Using SARIMAX with exogenous features like temperature and humidity gives more accurate forecasts.

🔹 3. Validation and Monitoring

* Split data into train/validation/test sets (e.g., rolling-window validation).

* Use metrics: RMSE, MAE, and MAPE for forecast accuracy.

* Monitor anomalies with thresholds (e.g., alert if usage deviates >20% from forecast).

* Deploy dashboards (Grafana, PowerBI) to continuously track both forecasts and anomalies.

🔹 4. Business Impact

This system helps:

* Prevent power outages by detecting sudden surges.

* Optimize resource allocation by forecasting demand region-wise.

* Support dynamic pricing (higher rates during peaks, lower during lows).

* Enable preventive maintenance if anomalies signal failing equipment.

```python
#Q10. Real-time workflow for energy demand anomaly detection and forecasting

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Simulated streaming energy usage data (15-minute intervals)
np.random.seed(42)
time_index = pd.date_range("2025-01-01", periods=200, freq="15min")
energy_usage = np.random.normal(100, 10, 200)

# Inject anomalies
energy_usage[50] = 200   # spike
energy_usage[120] = 20   # drop

df = pd.DataFrame({"time": time_index, "usage": energy_usage})

# ---- Step 1: Anomaly Detection ----
clf = IsolationForest(contamination=0.02, random_state=42)
df["anomaly"] = clf.fit_predict(df[["usage"]])

# Plot anomalies
plt.figure(figsize=(12,5))
plt.plot(df["time"], df["usage"], label="Energy Usage")
plt.scatter(df["time"][df["anomaly"]==-1], df["usage"][df["anomaly"]==-1],
            color="red", label="Anomaly")
plt.title("Real-time Energy Usage with Anomaly Detection")
plt.xlabel("Time")
plt.ylabel("Energy Usage (kWh)")
plt.legend()
plt.show()

# ---- Step 2: Short-term Forecasting (using SARIMA on 1-hour aggregated data) ----
df_resampled = df.resample("H", on="time").mean()

model = SARIMAX(df_resampled['usage'], order=(1,1,1), seasonal_order=(1,1,1,24))
results = model.fit(disp=False)

forecast = results.get_forecast(steps=24)
forecast_index = pd.date_range(start=df_resampled.index[-1]+pd.offsets.Hour(1), periods=24, freq="H")

# Plot forecast
plt.figure(figsize=(12,5))
plt.plot(df_resampled['usage'], label="Observed")
plt.plot(forecast_index, forecast.predicted_mean, color="red", label="Forecast")
plt.fill_between(forecast_index,
                 forecast.conf_int()["lower usage"],
                 forecast.conf_int()["upper usage"],
                 color="pink", alpha=0.3)
plt.title("Short-term Forecast of Energy Demand (SARIMA)")
plt.xlabel("Time")
plt.ylabel("Energy Usage (kWh)")
plt.legend()
plt.show()
```
